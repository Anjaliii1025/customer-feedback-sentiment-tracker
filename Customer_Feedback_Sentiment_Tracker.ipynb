```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Tweets.csv')
df = df[['text', 'airline_sentiment']]
df.dropna(inplace=True)
df.head()
```

```python
sns.countplot(data=df, x='airline_sentiment', palette='Set2')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
df['airline_sentiment'].value_counts().plot.pie(
    autopct='%1.1f%%',
    colors=['lightgreen', 'gold', 'lightcoral'],
    labels=['Negative', 'Neutral', 'Positive'],
    startangle=140
)
plt.title('Sentiment Breakdown')
plt.ylabel('')
plt.show()
```

```python
total = len(df)
positive_pct = round((df['airline_sentiment'] == 'positive').sum() / total * 100, 2)
negative_pct = round((df['airline_sentiment'] == 'negative').sum() / total * 100, 2)
neutral_pct = round((df['airline_sentiment'] == 'neutral').sum() / total * 100, 2)

print(f" Positive Feedback: {positive_pct}%")
print(f" Negative Feedback: {negative_pct}%")
print(f" Neutral Feedback: {neutral_pct}%")
```
